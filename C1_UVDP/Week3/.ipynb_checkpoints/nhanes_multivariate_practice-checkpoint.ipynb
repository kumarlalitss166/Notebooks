## Question 3

Use "violin plots" to compare the distributions of ages within groups defined by gender and educational attainment.

```python
print(da.RIDAGEYR.max())
print(da.RIDAGEYR.min())
```

```python
df['agegrp']=pd.cut(da.RIDAGEYR,[18,30,40,50,60,70,80])
df['Education']=da.DMDEDUC2.replace({1:'<9',2:'9-11',3:'HS/GED',4:'Some college/AA',5:'College',7:'Refused',9:'Don\'t know'})
df['RIDAGEYR']=da.loc[:,'RIDAGEYR']
df
```

```python
plt.figure(figsize=(3,5))
sns.violinplot(y='RIDAGEYR',x='RIAGENDRx',data=df)
```

```python
plt.figure(figsize=(16,5))
sns.violinplot(y='RIDAGEYR',x='Education',data=df)
```

```python
plt.figure(figsize=(16,5))
sns.violinplot(data=df,x='Education',y='RIDAGEYR',hue='RIAGENDRx')
```

__Q4a.__ Comment on any evident differences among the age distributions in the different demographic groups.

The key difference among the statrified violinplots is that distribution of people with eductation attainment `College` is left **slightly right-skewed** where distribution of people with education attainment `<9` is **slightly left skewed** and others have fairly normal distribution.

## Question 5

Construct a frequency table for the joint distribution of ethnicity groups ([RIDRETH1](https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.htm#RIDRETH1)) and health-insurance status ([HIQ210](https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/HIQ_I.htm#HIQ210)).
```python
da.loc[:,'HIQ210']
da.HIQ210.unique()
```

```python
df['HI']=da.HIQ210.replace({1:'Yes',2:'No',7:'Refused',9:'Don\'t know'})
df
```

```python
df.HI.unique()
```

```python
x=pd.crosstab(df.Ethnicity,df.HI)
x
```
